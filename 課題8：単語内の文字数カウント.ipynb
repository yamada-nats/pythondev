```python
word_list = []
alp = dict.fromkeys(list("abcdefghijklmnopqrstuvwxyz"), 0)

while True:
    word = input("英単語を入力してください：")
    if word == "":
        break
    word_list.append(word)
    for char in list(word):
        alp[char] += 1
    
print(f"入力した英単語：{sorted(word_list)}")

for char in alp.keys():
    if alp[char] > 0:
        print(f"{char}が{alp[char]}個ありました")
```
